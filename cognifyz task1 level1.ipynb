```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")  
df
```

```python
df_cuisine = df.dropna(subset=["Cuisines"])
```

```python
cuisine_series = df_cuisine["Cuisines"].str.split(",").explode().str.strip()
```

```python
top_cuisines = cuisine_series.value_counts().head(3)
```

```python
total_restaurants = df_cuisine.shape[0]
```

```python
top_cuisine_percentages = (top_cuisines / total_restaurants * 100).round(2)
```

```python
print("Top 3 Most Common Cuisines:\n", top_cuisines)   
print("\nPercentage of Restaurants Serving Each Top Cuisine:\n", top_cuisine_percentages)
```

```python
plt.figure(figsize=(8,5))
bars = plt.bar(top_cuisine_percentages.index, top_cuisine_percentages.values, color='skyblue')
plt.title("Top 3 Most Common Cuisines by Percentage")
plt.xlabel("Cuisine")
plt.ylabel("Percentage of Restaurants (%)")
```
