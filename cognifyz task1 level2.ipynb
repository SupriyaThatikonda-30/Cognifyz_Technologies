```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
df
```

```python
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
```

```python
df_rated = df.dropna(subset=['Aggregate rating'])
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df_rated['Aggregate rating'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
rounded_ratings = df_rated['Aggregate rating'].round(1)
most_common_rating = rounded_ratings.value_counts().idxmax()
print(f"Most common rating: {most_common_rating}")
```

```python
average_votes = df_rated['Votes'].mean()
print(f"Average number of votes received by restaurants: {average_votes:.2f}")
```
